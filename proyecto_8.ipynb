## Paso 5: Prueba de hipótesis



### Planteamiento de hipótesis
   
1. Hipótesis nula (H0):
La duración promedio del viaje en sábados lluviosos es igual a la duración promedio en sábados sin lluvia (o en condiciones buenas).

2. Hipótesis alternativa (H1):
La duración promedio cambia (es diferente) en sábados lluviosos respecto a sábados sin lluvia.


### Criterio para la prueba

Como comparamos la media de duración entre dos grupos independientes (viajes en sábados lluviosos vs sábados no lluviosos), usaremos una prueba t de Student para muestras independientes.

Nivel de significancia α típico: 0.05 (5%).


```python
import pandas as pd

# Cargar la ruta del archivo 
df = pd.read_csv('/datasets/project_sql_result_07.csv')

# Mostrar las primeras filas para verificar que se cargó correctamente
print(df.head())
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Asumiendo que ya tienes cargado el DataFrame 'df' con las columnas:
# 'start_ts' (fecha/hora), 'weather_conditions' ('Good' o 'Bad'), 'duration_seconds'

# Convertir start_ts a datetime y filtrar solo sábados
df['start_ts'] = pd.to_datetime(df['start_ts'])
df_saturdays = df[df['start_ts'].dt.dayofweek == 5].copy()  # 5 = sábado

# Separar grupos según condición climática: 'Bad' o 'Good'
durations_bad = df_saturdays[df_saturdays['weather_conditions'] == 'Bad']['duration_seconds'].dropna()
durations_good = df_saturdays[df_saturdays['weather_conditions'] == 'Good']['duration_seconds'].dropna()

print(f"Número de viajes sábados Bad: {len(durations_bad)}")
print(f"Número de viajes sábados Good: {len(durations_good)}")

# Realizar prueba t independiente
stat, p_value = ttest_ind(durations_bad, durations_good, equal_var=False)  # Welch's t-test

print(f'Estadístico t: {stat}')
print(f'Valor p: {p_value}')

# Interpretación del resultado para alfa = 0.05
if p_value < 0.05:
    print("Rechazamos la hipótesis nula: la duración promedio difiere entre sábados con clima 'Bad' y 'Good'.")
else:
    print("No rechazamos la hipótesis nula: no hay evidencia suficiente para decir que la duración promedio cambia con el clima.")
```

Bien hecho!<br/>

Es correcto rechazar la hipotesis nula, debido a que sí hay notables cambios entre un día de buenas condiciones y uno con malas condiciones
</div>

El valor p es extremadamente bajo (mucho menor que el nivel de significancia común α = 0.05), lo que indica que la diferencia observada en la duración promedio de los viajes entre sábados con clima "Bad" y "Good" es estadísticamente significativa.

Por lo tanto, rechazamos la hipótesis nula (que dice que no hay diferencia en la duración promedio de los viajes).

Esto sugiere que las condiciones climáticas adversas (mal tiempo) sí afectan la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O’Hare los sábados.


## Conclusion

El análisis realizado sobre los viajes desde el Loop hasta el Aeropuerto Internacional O’Hare en Chicago durante los sábados de noviembre de 2017 demuestra que las condiciones climáticas adversas, categorizadas como “Bad” (incluyendo lluvia y tormentas), tienen un impacto significativo en la duración promedio de los viajes.

La prueba estadística de hipótesis indicó una diferencia significativa en la duración de los viajes entre sábados con buen clima (“Good”) y sábados con mal clima (“Bad”), con un valor p muy inferior al nivel de significancia establecido (α = 0.05). Esto sugiere que en días de mal tiempo los viajes tienden a durar más, posiblemente debido a condiciones viales adversas o mayor congestión.
